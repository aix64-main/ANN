#**Perceptron**


###Any AI model can be considered as:

##**y = f(X)**

###For example in a perceptron model, f(X) is as shown below:
##**y = Ï•(WX+b)**

perceptron.png

Perceptron is a supervised method. Thus, we need a dataset of examples in the form of:

###**<data,lable> or <X,y>**

to train the model and estimate optimal values for W,b

###**How are the weights updated?**

**Every weight Wi is updated by Î”(Wi).**

Wi = Wi + Î”(Wi)

**Where:**

Î”(Wi) = Î· (y â€” Å·) * Xi

**And:**

Wi is i-th weight.

Î”(Wi) is the update. It determines if the weight increases or decreases.


Î· is the learning rate.


y is the actual value.


Å· is the predicted value.


Xi is the i-th input.

#**Generate Data**

```python
apple = 30
orange = 40

orders=[
    [7,2],
    [2,9],
    [5,5],
    [6,4],
    [6,3],
    [4,8],
    [1,7]
]

x_data=[]
y_data=[]

print('[apple,orange] -> total_price \n')
for ord in orders:
    total_price=(ord[0]*apple)+(ord[1]*orange)
    y_data.append(total_price)
    print(ord,'->',total_price)

x_data=orders



x_train=x_data[:5]
x_test=x_data[5:]

y_train=y_data[:5]
y_test=y_data[5:]

print('\n\n')
print('[apple,orange] -> total_price (Train Data)')
for x,y in zip(x_train,y_train):
    print(x,'->',y)

print('\n\n\n[apple,orange] -> total_price (Test Data)')
for x,y in zip(x_test,y_test):
    print(x,'->',y)

print('\n'*5)


w0=0
w1=0
b=0
lr=0.01

all_errors = []
for epoch in range(20):
    for x,y in zip(x_train,y_train):
        output = (w0*x[0])+(w1*x[1])+b
        error = (y-output)
        all_errors.append(error)
        delta_w0 = lr*error*x[0]
        delta_w1 = lr*error*x[1]
        w0 = w0+delta_w0
        w1 = w1+delta_w1
        b = b+(lr*error)
    if epoch%5==0:
        print("epoch:",epoch,"error:",error,"\t w0:",w0,"\t w1:",w1,"\t b:",b)
        # lr/=2



import matplotlib.pyplot as plt

a_err = [abs(a) for a in all_errors]
# plt.plot(all_errors, color='r', label='error')
plt.plot(a_err, color='g', label='abs(error)')
plt.xlabel("epoch")
plt.ylabel("error")
plt.legend()
plt.show()


print("\n\n Results for test data:")
for x,y in zip(x_test,y_test):
    output = (w0*x[0])+(w1*x[1])+b
    error = (y-output)
    print("X:",x,"target:",y,"\t output:",output,"\t error:",error)
```

#**Example (Logical Gates AND,OR,XOR):**

```python
x_train=[[0,0],
         [0,1],
         [1,0],
         [1,1]]

y_train=[0,1,1,1]  #OR
# y_train=[0,0,0,1]  #AND
# y_train=[0,1,1,0]  #XOR

for x,y in zip(x_train,y_train):
    if y==0:
        plt.plot(x[0],x[1],color='r',marker='+',label='zero')
    else:
        plt.plot(x[0],x[1],color='b',marker='^',label='one')
plt.xlabel('x0')
plt.ylabel('x1')
plt.axis([-1,2,-1,2])
plt.legend()
plt.show()
```

```python
w0, w1, b = 0., 0., -1

def activate(x):
    return 1 if x >= 0 else 0

def train_perceptron(inputs, desired_outputs, learning_rate, epochs):
    global w0, w1, b
    for epoch in range(epochs):
        total_error = 0
        for i in range(len(inputs)):
            A, B = inputs[i]
            target_output = desired_outputs[i]
            output = activate(w0 * A + w1 * B + b)
            error = target_output - output
            w0 += learning_rate * error * A
            w1 += learning_rate * error * B
            b += learning_rate * error
            total_error += abs(error)
        if total_error == 0:
            break


inputs = x_train
desired_outputs = y_train
learning_rate = 0.1
epochs = 100

train_perceptron(inputs, desired_outputs, learning_rate, epochs)

for i in range(len(inputs)):
    A, B = inputs[i]
    output = activate(w0 * A + w1 * B + b)
    print(f"Input: ({A}, {B})  Output: {output}")

print("w0:",w0,"w1:",w1,"b:",b)
```
